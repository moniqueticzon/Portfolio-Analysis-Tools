# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
whale_data = Path("Resources/whale_returns.csv")
# Reading whale returns
whale_df = pd.read_csv(whale_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
whale_df.head()
```

```python
whale_df.shape
```

```python
# Count nulls
whale_df.isna().sum()
```

```python
# Drop nulls
whale_returns_df = whale_df.dropna()
whale_returns_df.isna().sum()
```

```python
whale_returns_df.shape
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
algo_data = Path("Resources/algo_returns.csv")
# Reading algorithmic returns
algo_df = pd.read_csv(algo_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
algo_df.head()
```

```python
# Count nulls
algo_df.isna().sum()
```

```python
# Drop nulls
algo_returns_df = algo_df.dropna()
algo_returns_df.isna().sum()
```

```python
algo_returns_df.head()
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
sp500_data = Path("Resources/sp500_history.csv")
# Reading S&P 500 Closing Prices
sp500_df = pd.read_csv(sp500_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp500_df.head()
```

```python
sp500_df.shape
```

```python
sp500_df[sp500_df.columns] = sp500_df[sp500_df.columns].replace({'\$':''}, regex = True)
print (sp500_df)
```

```python
# Count nulls
sp500_df.isna().sum()
```

```python
# Drop nulls
sp500_returns_df = sp500_df.dropna()
sp500_returns_df.isna().sum()
```

```python
sp500_returns_df.head()
```

```python
# Check Data Types
sp500_returns_df.dtypes
```

```python
# Fix Data Types
sp500_returns_df["Close"] = sp500_returns_df["Close"].astype(float).astype(int)
print(sp500_returns_df.dtypes)
```

```python
sp500_returns_df.astype('int64').dtypes
```

```python
# Calculate Daily Returns
sp500_daily_returns_df = sp500_returns_df.pct_change()
sp500_daily_returns_df.head()
```

```python
# Drop nulls
sp500_daily_returns_clean = sp500_daily_returns_df.dropna()
sp500_daily_returns_clean.head()
```

```python
sp500_daily_returns_clean.isna().sum()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp500_daily_returns_clean.rename(columns = {'Close': 'S&P 500'}, inplace = True)
```

```python
sp500_daily_returns_clean.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
combined_df = pd.concat([whale_returns_df, algo_returns_df, sp500_daily_returns_clean], axis="columns", join="inner")
combined_df.sort_index(inplace=True)
combined_df
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta.

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
rolling_std_21 = combined_df.rolling(window=21).mean()

# Plot the rolling standard deviation
rolling_std_21.plot(figsize=(20,15), title="Rolling 21-Day STD of All Portfolios")
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation = combined_df.corr()
# Display de correlation matrix
correlation
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
covariance_algo_1 = combined_df['Algo 1'].cov(combined_df['S&P 500'])
# Calculate variance of S&P 500
variance = combined_df['S&P 500'].var()
# Computing beta
algo_1_beta = covariance_algo_1 / variance
algo_1_beta
```

```python
# 60-day rolling beta between Algo 1 and S&P 500
# Calculate covariance of a single portfolio
rolling_covariance = combined_df['Algo 1'].rolling(window=60).cov(combined_df['S&P 500'])
# Calculate variance of S&P 500
rolling_variance = combined_df['S&P 500'].rolling(window=60).var()
# Plot beta trend on 60-day rolling beta
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20,15), title="Rolling 60-Day Beta of Algo 1")
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratios = (combined_df.mean() * 252) / (combined_df.std() * np.sqrt(252))
sharpe_ratios.sort_values(ascending=False)
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

```python
# Reading data from 1st stock
orcl_data = Path("Resources/orcl_data.csv")
orcl_df = pd.read_csv(orcl_data, index_col='Date', parse_dates=True, infer_datetime_format=True)
orcl_df.head()
```

```python
# Reading data from 2nd stock
rlgt_data = Path("Resources/rlgt_data.csv")
rlgt_df = pd.read_csv(rlgt_data, index_col='Date', parse_dates=True, infer_datetime_format=True)
rlgt_df.head()
```

```python
# Reading data from 3rd stock
xec_data = Path("Resources/xec_data.csv")
xec_df = pd.read_csv(xec_data, index_col='Date', parse_dates=True, infer_datetime_format=True)
xec_df.head()
```

```python
# Combine all stocks in a single DataFrame
my_portfolio = pd.concat([orcl_df, rlgt_df,xec_df], axis="columns",join="inner")
my_portfolio.head()
```

```python
# Reset Date index
my_portfolio.sort_index(inplace=True)
```

```python
# Reorganize portfolio data by having a column per symbol
my_portfolio.columns = ['ORCL', 'RLGT', 'XEC']
my_portfolio.head()
```

```python
# Calculate daily returns
daily_returns = my_portfolio.pct_change()

# Drop NAs
daily_returns = daily_returns.dropna()

# Display sample data
daily_returns.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3]

# Calculate portfolio return
my_portfolio_returns = daily_returns.dot(weights)

# Display sample data
my_portfolio_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
all_portfolios_df = pd.concat([combined_df,my_portfolio_returns], axis="columns", join ="inner")
all_portfolios_df.sort_index(inplace=True)
all_portfolios_df.columns = ['SOROS FUND MANAGEMENT LLC', 'PAULSON & CO.INC.', 'TIGER GLOBAL MANAGEMENT LLC', 'BERKSHIRE HATHAWAY INC', 'Algo 1', 'Algo 2', 'S&P 500', 'My Portfolio']
all_portfolios_df.tail()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
all_portfolios_df.isna().sum()
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized `std`
annualized_std = all_portfolios_df.std() * np.sqrt(252)
annualized_std.sort_values(ascending=False)
```

```python
# Calculate rolling standard deviation
rolling_std_21_all = all_portfolios_df.rolling(window=21).mean()

# Plot the rolling standard deviation
rolling_std_21_all.plot(figsize=(20,15), title="Rolling 21-Day STD of All Portfolios")
```

### Calculate and Plot Rolling 60-day Beta for Your Portfolio compared to the S&P 500

```python
# Calculate and plot Beta
covariance_my_portfolio = all_portfolios_df['My Portfolio'].cov(all_portfolios_df['S&P 500'])
variance_my_portfolio = all_portfolios_df['S&P 500'].var()
my_beta = covariance_my_portfolio / variance_my_portfolio
my_beta
```

```python
# 60-day rolling beta between My Portfolio and S&P 500
my_rolling_covariance = all_portfolios_df['My Portfolio'].rolling(window=60).cov(all_portfolios_df['S&P 500'])
my_rolling_variance = all_portfolios_df['S&P 500'].rolling(window=60).var()
# Plot beta trend on 60-day rolling beta
my_rolling_beta = my_rolling_covariance / my_rolling_variance
my_rolling_beta.plot(figsize=(20,15), title="Rolling 60-Day Beta of My Portfolio")
```
